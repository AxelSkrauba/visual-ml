**Herramienta interactiva** para explorar cómo los modelos clásicos de Machine Learning trazan fronteras de decisión.

Desarrollada para cursos de IA en ingeniería, permite visualizar de forma intuitiva conceptos como:
- **Overfitting / Underfitting**
- **Complejidad del modelo vs. generalización**
- **Impacto de hiperparámetros** en la forma de la frontera
- **Comparación simultánea** de múltiples algoritmos
- **Desbalance de clases** y su efecto en las métricas

### Modelos incluidos
Logistic Regression · Perceptron · SVC (lineal/RBF/poly) · KNN · Decision Tree · Random Forest · Gradient Boosting · AdaBoost · Gaussian NB · QDA

### Datasets
Moons · Circles · Blobs · XOR · Linear · Spirals · Desbalanceado

## Celda 4 — Lanzar la aplicación (sin ngrok, sin token externo)

Usa el port forwarding nativo de Google Colab.
El panel interactivo aparece **embebido directamente abajo** de esta celda.

````python
import subprocess
import threading
import time

PORT = 8501

# Kill any existing streamlit process
import os
os.system("pkill -f streamlit 2>/dev/null || true")
time.sleep(1)

# Launch Streamlit in background
def _run():
    subprocess.run(
        ["streamlit", "run", "app.py",
         "--server.port", str(PORT),
         "--server.headless", "true",
         "--server.enableCORS", "false",
         "--server.enableXsrfProtection", "false"],
        stdout=subprocess.DEVNULL,
        stderr=subprocess.DEVNULL,
    )

threading.Thread(target=_run, daemon=True).start()
time.sleep(4)  # wait for Streamlit to be ready

# Use Colab's native port forwarding — no external token needed
from google.colab.output import serve_kernel_port_as_iframe
serve_kernel_port_as_iframe(PORT, height=800)
````

## ℹ️ Notas

`google.colab.output.serve_kernel_port_as_iframe` es la forma **oficial y nativa** de Colab
para exponer un servidor local directamente como iframe en el notebook.
No requiere cuenta externa, token ni configuración adicional. En caso de necesidad, se puede usar ngrok para tunneling.
